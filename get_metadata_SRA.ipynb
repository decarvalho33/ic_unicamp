```python
import numpy as np
import requests

from google.colab import drive
drive.mount('/content/drive')
import csv
```

```python
ids = []

with open('ids.csv', 'r') as file:
    reader = csv.reader(file)

    for line in reader:
        column1 = line[0]
        ids.append(column1[4:])
```

```python
urls = []

for id in ids:
  urls.append('https://www.ncbi.nlm.nih.gov/sra?LinkName=biosample_sra&from_uid='+str(id))
```

Importamos algumas bibliotecas

```python
all_text = []

#text = requests.get(urls[0]).content.decode('utf-8')

i = 0
for url in urls:
    text = requests.get(url).content.decode('utf-8')
    all_text.append(text)
```

```python
print(len(all_text))
```

```python
text = all_text[0]


bases_index_start = (text.index("spots, "))+7
bases_index_end = text[bases_index_start:bases_index_start+20].rfind("bases")

print(text[bases_index_start:bases_index_start+bases_index_end])
```

```python
runs = []
spots = []
bases = []
i = 0
indexes = []

run_len = 11
threshold = 30

#extrair numero RUN
for text in all_text:

  run_index = text.index("run=")+4
  spots_index_start = (text.index("run: "))+5
  bases_index_start = (text.index("spots, "))+7

  runs.append(text[(run_index):(run_index+run_len)]) #funcionando

  spots_index_end = text[spots_index_start:spots_index_start+threshold].rfind("spots")
  spots.append(text[spots_index_start:spots_index_start+spots_index_end])

  bases_index_end = text[bases_index_start:bases_index_start+threshold].rfind("bases")
  bases.append(text[bases_index_start:bases_index_start+bases_index_end])

  indexes.append(i)

  i += 1
```

```python
print("run: \n")
for r in runs:
  print(r)

print("\nspots:\n")
for sp in spots:
  print(sp)

print("\nbases: \n")
for bs in bases:
  print(bs)
```
